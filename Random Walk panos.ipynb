```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=200 #poses fores tha ginei to bhma
p=1
p1=1
p2=1
P=[]
pstay=0
pm=0
pp=0
x=0
pos=[0]
rand=[]
k=0
for i in range(0,N):
    r = np.random.rand()
    rand+=[r]
    if r<0.5: 
        x=x-1
        pm += 1 #vhmata pou ekane pros ta aristera 
    if r>0.5: 
        x=x+1
        pp += 1 #vhmata pou ekane pros ta dexia
    pos+=[x]  ###edw tha ithela na vrw posi pithanotita exei kathe simeio. alla mperdeumoa dioti einai idietera periploko 
    if abs(x) > k :
        p1=p1*1/2
        P+=[p1]
    elif abs(x) < k :
        p2=(k*((1/2)**k))
        P+=[p2]
    else :
        pstay=0
    k+=1
   
   
print(x,"\n")
print("ekane",pm,"vimata pisw",", kai",pp,"vimata mprosta")
xpos=np.array(pos)
plt.plot(xpos)
plt.show()
print("\n",pos,"\n\n", P)
```
