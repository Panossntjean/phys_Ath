```python
import math
import numpy as np
import matplotlib.pyplot as plt

N = 10000

xvals = np.linspace(0,4*math.pi,N )
h = (4*math.pi)/(N*0.1)

def f(x):
    val = math.sin(x)
    return val

     
def df(x):
    dval = (f(x+h)-f(x))/h
    return dval

yvals = []

for i in xvals:
    der = df(i)
    yvals += [der]
    
fvals = [] 

for i in xvals:
    sin = f(i)
    fvals += [sin]
    
plt.plot(xvals,yvals)
plt.plot(xvals,fvals, 'r')
plt.grid()
plt.title("Ploting the Derivative of sinx")
plt.show
    
```

```python
def f(x):
    val = math.cos(x)
    return val
fcvals=[]

for i in xvals:
    cos = f(i)
    fcvals += [cos]
    
plt.plot(xvals,fvals)
plt.plot(xvals,fcvals,'g')
plt.grid()
plt.show
```
