```python
import numpy as np 
import matplotlib.pyplot as plt

```

```python
N1 = 1000 #arxiki timi moriwn prwto doxeio 
N2 = 0    #arxiki timi moriwn detero doxeio
p = 1/100 #pithanotita metavasis moriou apo to ena sto allo doxeio 
T = 500  #poses fores tha trexoume to peirama

N1t = []
N2t = []
N1t+=[N1] #arithmos moriwn sto archiko doxeio me ton chrono  
N2t+=[N2] #arithmos moriwn sto deutero doxeio me ton chrono

#PEIRAMA

for i in range(0,T):
    d12 = 0
    d21 = 0
    
    for j in range (0,N1):
        if(np.random.rand()<p):
            d12+=1
        
    for j in range(0,N2):
        if(np.random.rand() < p):
            d21+=1   
    N1 += d21-d12
    N2 += d12-d21
    N1t+=[N1]
    N2t+=[N2]
    

    
plt.plot(N1t)+plt.plot(N2t)
```
